```python
def extract_area(paket):
    if "Sukabumi" in paket or "SKMI" in paket or "SMI" in paket or "Sukabum" in paket:
        return "Sukabumi"
    elif "Bandung" in paket or "Paskal" in paket or "Hotspot" in paket:
        return "Bandung"
    # Tambahkan kondisi untuk area lainnya sesuai kebutuhan
    else:
        return "Jakarta"  # Jika tidak ada area yang dikenali

# Terapkan fungsi ke kolom "Paket Langganan" dan buat kolom baru "Area"
all_data['Area'] = all_data['Paket_Langganan'].apply(extract_area)

# Tampilkan beberapa baris setelah pemisahan
print(all_data[['Paket_Langganan', 'Area']].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengubah Tanggal_Aktif menjadi format tanggal
all_data['Tanggal_Aktif'] = pd.to_datetime(all_data['Tanggal_Aktif'], errors='coerce')

# Cek apakah ada nilai yang hilang setelah konversi
print(all_data['Tanggal_Aktif'].isnull().sum())

# Mengambil bulan dan tahun
all_data['Bulan_Tahun'] = all_data['Tanggal_Aktif'].dt.to_period('M')

# Menghitung total pendapatan per bulan
pendapatan_per_bulan = all_data.groupby('Bulan_Tahun')['Total'].sum().reset_index()

# Pastikan 'Bulan_Tahun' diubah menjadi string untuk visualisasi
pendapatan_per_bulan['Bulan_Tahun'] = pendapatan_per_bulan['Bulan_Tahun'].astype(str)

# Visualisasi
plt.figure(figsize=(12, 6))
sns.lineplot(x='Bulan_Tahun', y='Total', data=pendapatan_per_bulan, marker='o', color='blue')

# Menambahkan label nominal ke setiap titik
for index, value in enumerate(pendapatan_per_bulan['Total']):
    plt.text(index, value, f'{value:,.0f}', ha='center', va='bottom', fontsize=10)

plt.title('Tren Pendapatan per Bulan', fontsize=16)
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Total Pendapatan', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong
plt.show()
```

```python
# Pastikan kolom 'Tanggal_Aktif' diubah ke format datetime
all_data['Tanggal_Aktif'] = pd.to_datetime(all_data['Tanggal_Aktif'])

# Buat kolom 'Bulan_Tahun' untuk pengelompokan
all_data['Bulan_Tahun'] = all_data['Tanggal_Aktif'].dt.to_period('M')

# Hitung total profit per bulan
profit_per_bulan = all_data.groupby('Bulan_Tahun').agg({
    'Total': 'sum',
    'PPN': 'sum',
    'Fee_Seller': 'sum'
}).reset_index()

# Hitung profit
profit_per_bulan['Profit'] = profit_per_bulan['Total'] - profit_per_bulan['PPN'] - profit_per_bulan['Fee_Seller']

# Visualisasi
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x='Bulan_Tahun', y='Profit', data=profit_per_bulan, color='lightgreen')
plt.title('Profit Perusahaan per Bulan', fontsize=16)
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()

# Tambahkan label nominal di atas batang
for p in barplot.patches:
    barplot.annotate(f'Rp {p.get_height():,.0f}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='baseline', fontsize=12, color='black', 
                     xytext=(0, 5), 
                     textcoords='offset points')

plt.show()

# Tampilkan profit per bulan
print(profit_per_bulan[['Bulan_Tahun', 'Profit']])
```

```python
import pandas as pd
from datetime import datetime

# Ubah 'Tanggal_Aktif' menjadi datetime
all_data['Tanggal_Aktif'] = pd.to_datetime(all_data['Tanggal_Aktif'])

# Hitung Recency
current_date = datetime.now()
rfm = all_data.groupby('ID_Pelanggan').agg({
    'Tanggal_Aktif': lambda x: (current_date - x.max()).days,
    'Invoice': 'count',
    'Total': 'sum'
}).reset_index()
rfm.columns = ['ID_Pelanggan', 'Recency', 'Frequency', 'Monetary']

# Tampilkan hasil RFM
print(rfm.head())

# Skor RFM (skala 1-5)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=False, duplicates='drop') + 1
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=False, duplicates='drop') + 1
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=False, duplicates='drop') + 1

# Klasifikasi
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Tampilkan hasil akhir
print(rfm[['ID_Pelanggan', 'Recency', 'Frequency', 'Monetary', 'RFM_Score']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Histogram untuk Recency
plt.figure(figsize=(12, 6))
sns.histplot(rfm['Recency'], bins=30, kde=True)
plt.title('Distribusi Recency', fontsize=16)
plt.xlabel('Recency (Hari)', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Histogram untuk Frequency
plt.figure(figsize=(12, 6))
sns.histplot(rfm['Frequency'], bins=30, kde=True)
plt.title('Distribusi Frequency', fontsize=16)
plt.xlabel('Frequency (Jumlah Transaksi)', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Histogram untuk Monetary
plt.figure(figsize=(12, 6))
sns.histplot(rfm['Monetary'], bins=30, kde=True)
plt.title('Distribusi Monetary', fontsize=16)
plt.xlabel('Monetary (Total Pengeluaran)', fontsize=14)
plt.ylabel('Jumlah Pelanggan', fontsize=14)
plt.tight_layout()
plt.show()

# Heatmap untuk skor RFM
plt.figure(figsize=(8, 6))
rfm_heatmap = rfm.pivot_table(index='R_Score', columns='F_Score', values='ID_Pelanggan', aggfunc='count', fill_value=0)
sns.heatmap(rfm_heatmap, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.title('Heatmap Jumlah Pelanggan Berdasarkan RFM Score', fontsize=16)
plt.xlabel('F_Score', fontsize=14)
plt.ylabel('R_Score', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Menghitung pendapatan bersih per bulan
monthly_income_df = all_data.resample(rule='M', on='Tanggal_Aktif').agg({
    "Total": "sum",
    "PPN": "sum",
    "Fee_Seller": "sum"
})

# Menambahkan kolom Pendapatan Bersih
monthly_income_df['Pendapatan_Bersih'] = (
    monthly_income_df['Total'] - monthly_income_df['PPN'] - monthly_income_df['Fee_Seller']
)

# Format indeks ke format 'YYYY-MM'
monthly_income_df.index = monthly_income_df.index.strftime('%Y-%m')

# Reset index untuk keperluan visualisasi
monthly_income_df = monthly_income_df.reset_index()
```

```python
monthly_order_df = all_data.resample(rule='M', on='Tanggal_Aktif').agg({
    "Invoice": "nunique",
    "Pendapatan_Bersih": "sum"
})

monthly_order_df.index = monthly_order_df.index.strftime('%Y-%m')
monthly_order_df = monthly_order_df.reset_index ()
monthly_order_df.rename(columns={
    "Invoice": "order_count",
    "Pendapatan_Bersih": "revenue"
}, inplace=True)
monthly_order_df.head() 
```
